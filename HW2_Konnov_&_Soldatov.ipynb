### Agricultural raw materials exports (% of merchandise exports)

**H1:** The distribution of Agricultural raw materials exports (% of merchandise exports) is different from the normal distribution

**H0:** The distribution of Agricultural raw materials exports (% of merchandise exports) variable is not different from the normal distribution

```python
column_name = 'Agricultural raw materials exports (% of merchandise exports)'

kstest_result = stats.kstest(df[column_name].dropna(), 'norm', args=(df[column_name].dropna().mean(), df[column_name].dropna().std()))
p_value = kstest_result.pvalue

print(kstest_result)
```

```python
import seaborn as sns
from scipy.stats import norm
sns.histplot(df['Agricultural raw materials exports (% of merchandise exports)'].dropna(), kde=True, stat="density")
```

```python
import pingouin as pg
pg.qqplot(df['Agricultural raw materials exports (% of merchandise exports)'], dist = 'norm')
```

```python
if p_value < 0.05:
    print(f"Conclusion: p-value = {p_value} => as p-value < 0.05, H1 is accepted => the distribution is significantly different from normal")
else:
    print(f"Conclusion: p-value = {p_value} => as p-value > 0.05, H1 is rejected => the distribution is not significantly different from normal")
```

## Task #3

Calculate appropriate correlation coefficients between any three pairs of variables. Explain
the selection of the correlation coefficient. Fill in the table below. Interpret the results. Create
suitable graphs to visualize the analyzed relationships.

### Exports of goods and services (current US$) & Merchandise exports (current US\$)

 **Pair of variables:** Exports of goods and services (current US$) & Merchandise exports (current US\$)

 **Applied method:** Spearman's rank correlation coefficient.

>  This method was chosen because both variables have a non-normal distribution as well as a small number of duplicates.



```python
jitter_strength = 1e10
merchandise_with_jitter = df['Merchandise exports (current US$)'] + np.random.uniform(-jitter_strength, jitter_strength, size=len(df))
exports_with_jitter = df['Exports of goods and services (current US$)'] + np.random.uniform(-jitter_strength, jitter_strength, size=len(df))

plt.figure(figsize=(8, 6))
plt.scatter(merchandise_with_jitter, exports_with_jitter, alpha=0.7, color='blue', s=10)
plt.title('Scatterplot of Merchandise exports vs. Exports of goods and services with Jitter')
plt.xlabel('Merchandise exports (current US$)')
plt.ylabel('Exports of goods and services (current US$)')
plt.grid(True)
plt.show()
```

```python
for column in ['Exports of goods and services (current US$)', 'Merchandise exports (current US$)']:
    unique_count = df[column].nunique()
    total_count = len(df[column])
    unique_percentage = (unique_count / total_count) * 100
    duplicate_percentage = 100 - unique_percentage

    print(f"{column}: {unique_percentage:.2f}% unique, {duplicate_percentage:.2f}% duplicates")
```

```python
df1 = df.dropna(subset=['Merchandise exports (current US$)', 'Exports of goods and services (current US$)'])
stats.spearmanr(df1['Merchandise exports (current US$)'], df1['Exports of goods and services (current US$)'])
```

```python
df[['Merchandise exports (current US$)','Exports of goods and services (current US$)']].corr('spearman')
```

```python
sns.regplot(data = df1, x = 'Merchandise exports (current US$)', y = 'Exports of goods and services (current US$)')
```

***Conclusion:***

**Variables:** Merchandise exports (current US\$) and Exports of goods and services (current US\$).

**Appropriate correlation coefficient**: Spearman's rank correlation coefficient.

**Hypotheses:**

### ECD_members and European_Union_members


**Pair of variables:** OECD_members and European_Union_members.

**Applied method:** Phi correation.

> This method was chosen because both variables are binary categorical variables.

```python
def phi_corrrelation(data: pd.DataFrame, x: str, y: str):
    chi = pd.crosstab(index=data[x],
                           columns=data[y], margins = True).values
    phi = (chi[0, 0] * chi[1, 1] - chi[1, 0] * chi[0, 1]) / np.sqrt(chi[1, 2] * chi[2, 1] * chi[2, 0] * chi[0, 2])
    return phi
```

```python
phi_corrrelation(df, 'OECD_members', 'European_Union_members')
```

Calculate a paired correlation coefficient between any variables. Then calculate the partial
correlation coefficient between the same pair of variables controlling for any other third
variable. Interpret the results of analysis. Create suitable graphs to visualize the analyzed
relationships.

Let's take two variables of Commercial service exports (current US\$) and Goods exports (BoP, current US\$) and see how do they correlate. We will control their correlation with a variable of Income Group

```python
import pingouin as pg

income_mapping = {
    'low income': 4,
    'lower middle income': 3,
    'upper midddle income': 2,
    'high income': 1
}

df['income_group_numeric'] = df['income_group'].map(income_mapping)
```

```python
x = df['Commercial service exports (current US$)'].replace([np.inf, -np.inf], np.nan).dropna()
y = df['Goods exports (BoP, current US$)'].replace([np.inf, -np.inf], np.nan).dropna()

pearson_corr, p_value = stats.pearsonr(x, y)

print(f"Pearson Correlation: {pearson_corr}")
print(f"P-value: {p_value}")
```

```python
partial_corr = pg.partial_corr(data=df,
                               x='Commercial service exports (current US$)',
                               y='Goods exports (BoP, current US$)',
                               covar='income_group_numeric')
print(partial_corr)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Commercial service exports (current US$)',
    y='Goods exports (BoP, current US$)',
    data=df
)
plt.title('Scatter Plot - Commercial Service Exports vs. Goods Exports')
plt.xlabel('Commercial service exports (current US$)')
plt.ylabel('Goods exports (BoP, current US$)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(
    x='income_group_numeric',
    y='Goods exports (BoP, current US$)',
    data=df
)
plt.title('Box Plot - Goods Exports vs. Income Groups')
plt.xlabel('Income Group')
plt.ylabel('Goods exports (BoP, current US$)')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(
    x='income_group_numeric',
    y='Commercial service exports (current US$)',
    data=df
)
plt.title('Box Plot - Commercial Service Exports vs. Income Groups')
plt.xlabel('Income Group')
plt.ylabel('Commercial service exports (current US$)')
plt.show()
```

The r coefficient has increased in partial test, and the p-value is insanely small, so we can come to conclusion that there's a correlation and third value enforces it.
